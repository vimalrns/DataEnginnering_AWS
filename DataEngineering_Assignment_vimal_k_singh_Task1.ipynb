```python
import pymongo 
import math

# define the client, the database, and the collection
client = pymongo.MongoClient('localhost',27017)
database = client["ftfinanace"]
collection = database["ftdata"]
```

```python
collection.find_one()
```

```python
import pandas as pd
datasetfin = collection.find()
entries = list(datasetfin)
entries[:5]
```

```python
df = pd.DataFrame(entries)
df.head()
```

```python
df.describe()
```

```python
# Loan defauld by emplyeement type
df.query("loan_default == '1'").groupby(["Employment_Type"])["disbursed_amount"].mean().reset_index()
```

```python
#Branch where the loan was disbursed with emplyeement type and PERFORM_CNS_SCORE=0
df.query("PERFORM_CNS_SCORE == '0'").groupby(["Employment_Type","branch_id"])["disbursed_amount"].max().reset_index()
```

```python
#Total disbursed amount by branch: This query summarizes the disbursed amount of loans by branch. 
#It provides an overview of the total loan amount disbursed by each branch.
# Grouping the data by branch_id and calculating the sum of disbursed_amount for each branch
branch_disbursed_amount = df.groupby('branch_id')['disbursed_amount'].sum()
branch_disbursed_amount
```

```python
# Loan default rate by age group: This query analyzes the loan default rate based on the age group of the borrowers. 
#It helps identify if certain age groups have a higher tendency to default on loans.

data=pd.DataFrame(df)
data['Date_of_Birth'] = pd.to_datetime(df['Date_of_Birth'])
data.dropna(subset=['Date_of_Birth'], inplace = True)
#data['Date_of_Birth'] = pd.to_datetime(data['Date_of_Birth'], format='%Y%m%d')
age_bins = [18, 30, 40, 50, 60, 100]  # Specify the age group boundaries
age_labels = ['18-29', '30-39', '40-49', '50-59', '60+']  # Labels for age groups

data['Age_Group'] = pd.cut(data['Date_of_Birth'].dt.year, bins=age_bins, labels=age_labels)

# Calculate loan default rate by age group
default_rate_by_age = data.groupby('Age_Group')['loan_default'].mean()

# Print the loan default rate by age group
default_rate_by_age

```

```python
#Employment type distribution: This query provides a breakdown of the borrowers' employment types, 
#allowing you to understand the distribution of loan applicants based on their employment status.
# Grouping the data by employment type and calculating the count
employment_type_distribution = df.groupby('Employment_Type').size()

# Calculating the percentage distribution
total_count = len(df)
employment_type_percentage = (employment_type_distribution / total_count) * 100

# Creating a new DataFrame to store the results
result_df = pd.DataFrame({
    'Employment_Type': employment_type_distribution.index,
    'Count': employment_type_distribution.values,
    'Percentage': employment_type_percentage.values
})

# Sorting the DataFrame by count in descending order
result_df = result_df.sort_values('Count', ascending=False)
result_df
```

```python
#Credit score analysis: This query examines the relationship between the borrowers'
#credit scores and their loan default status. It helps assess the creditworthiness of borrowers based on their credit history.
# Select relevant columns for analysis
columns_of_interest = ['PERFORM_CNS_SCORE', 'loan_default']
df_selected = df[columns_of_interest]

# Group data by credit score and calculate default rate
credit_score_analysis = df_selected.groupby('PERFORM_CNS_SCORE')['loan_default'].mean().reset_index()

# Sort the results by credit score in ascending order
credit_score_analysis = credit_score_analysis.sort_values('PERFORM_CNS_SCORE')
credit_score_analysis
```

```python
#Loan inquiries and default correlation:
#This query investigates the correlation between the number of loan inquiries made by borrowers and their default status. 
#It helps determine if there is a relationship between the borrowers' loan inquiry activities and their likelihood of defaulting on loans.
# Selecting the relevant columns for analysis
columns_of_interest = ['NO_OF_INQUIRIES', 'loan_default']
subset_data = data[columns_of_interest]

# Calculating the correlation
correlation = subset_data['NO_OF_INQUIRIES'].corr(subset_data['loan_default'])

# Printing the correlation coefficient
print("Correlation between Loan Inquiries and Loan Default: ", correlation)
```
